```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import user_data_analytics
```

```python
ld = list()
it = 0
while(True):
    try:
        data = pd.read_json(r'my_spotify_data\MyData\endsong_' + str(it) + '.json')
        ld.append(data)
        it += 1
    except:
        break
data = ld[0]
for i in range(1, len(ld)):
    data = pd.concat([data, ld[i]])
```

```python
ob = user_data_analytics.Analytics(data)
```

# 1) Listening time overall and yearly

```python
a, b = ob.listening_time()
```

```python
a
```

```python
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
sns.barplot(data=b, x='Year', y="Total_listening_time_in_ms")
plt.title('Listening Time Each Year')
plt.show()
```

```python
b.to_csv('Spotify_stats/listening_time_yearly.csv')
```
