```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

# 1)	List of favorite artists based on
## a.	Total number of songs listened to
## b.	Total listening time of the artist
## c.	Number of times songs of the artist have been played


```python
tot_artist_overall = data['master_metadata_album_artist_name'].unique()
artist_listentime_overall = list()
artist_totsongs_overall = list()
artist_songsplayed_overall = list()

# tot_artist_overall contains the list of all the artists listened to till now
# artist_listentime_overall is total listening time for an artist 
# artist_totsongs_overall is the total number of songs played by an artist doesn't include duplicates
# artist_songsplayed_overall is similar but includes duplicates too 
```

```python
for i in range(0, len(tot_artist_overall)):
    temp = data[data['master_metadata_album_artist_name'] == tot_artist_overall[i]]
    artist_listentime_overall.append(temp['ms_played'].sum())
    artist_totsongs_overall.append(temp['master_metadata_track_name'].unique().shape[0])
    artist_songsplayed_overall.append(temp.shape[0])
```

```python
df = pd.DataFrame(list(zip(tot_artist_overall, artist_listentime_overall, artist_totsongs_overall, artist_songsplayed_overall)), columns=['Artist', 'Total_listening_time_in_ms', 'Total_no_of_songs_listened', 'No_of_times_played'])
```

```python
df_favartist_listeningtime = df.sort_values(by=['Total_listening_time_in_ms'], ascending=False).drop(['Total_no_of_songs_listened', 'No_of_times_played'], axis=1).iloc[:50]
df_favartist_listeningtime.index = range(0, 50)
```

# 5)	Favorite time of the day for listening based on
## a.	Total number of songs listened to
## b.	Total listening time


```python
morn = [dt.time(6, 0, 0), dt.time(11, 59, 59)]
afnoon = [dt.time(12, 0, 0), dt.time(17, 59, 59)]
eve = [dt.time(18, 0, 0), dt.time(23, 59, 59)]
night = [dt.time(0, 0, 0), dt.time(5, 59, 59)]
time_of_day = [morn, afnoon, eve, night]
time_of_day_listentime = list()
time_of_day_songsplayed = list()
td = ['Morning', 'Afternoon', 'Evening', 'Night']
```

```python
for i in range(0,4):
    temp = data[[d >= time_of_day[i][0] and d <= time_of_day[i][1] for d in data['time']]]
    time_of_day_songsplayed.append(temp.shape[0])
    time_of_day_listentime.append(temp['ms_played'].sum())
```

```python
df = pd.DataFrame(list(zip(td, time_of_day_listentime, time_of_day_songsplayed)), columns=['Time_of_day', 'Total_listening_time_in_ms', 'Total_songs_listened_to'])
```

```python
df_favtime_msplayed = df.sort_values(by=['Total_listening_time_in_ms'], ascending=False).drop(['Total_songs_listened_to'], axis=1)
df_favtime_msplayed.index = range(0, 4)
```

```python
df_favtime_msplayed
```

```python
df_favtime_songsplayed = df.sort_values(by=['Total_songs_listened_to'], ascending=False).drop(['Total_listening_time_in_ms'], axis=1)
df_favtime_songsplayed.index = range(0, 4)
```

# 6)	Favorite artist for each time of the day based on
## a.	Total number of songs listened to
## b.	Total listening time

```python
timeofday_artist_time = [['nothing', 0], ['nothing', 0], ['nothing', 0], ['nothing', 0]]
timeofday_artist_songsplayed = [['nothing', 0], ['nothing', 0], ['nothing', 0], ['nothing', 0]]
```

```python
for i in range(0,4):
    temp = data[[d >= time_of_day[i][0] and d <= time_of_day[i][1] for d in data['time']]]
    temp_art = temp['master_metadata_album_artist_name'].unique()
    for j in temp_art:
        temp2 = temp[temp['master_metadata_album_artist_name'] == j]
        if temp2.shape[0] > timeofday_artist_songsplayed[i][1]:
            timeofday_artist_songsplayed[i][0] = j
            timeofday_artist_songsplayed[i][1] = temp2.shape[0]
        if temp2['ms_played'].sum() > timeofday_artist_time[i][1]:
            timeofday_artist_time[i][0] = j
            timeofday_artist_time[i][1] = temp2['ms_played'].sum()
```

```python
df_timeofday_artist_tottime = pd.DataFrame(timeofday_artist_time, columns=['Artist', 'Total_time_listened_in_ms'])
df_timeofday_artist_tottime.insert(loc=0, column='Time_of_day', value=td)
```

```python
df_timeofday_artist_tottime
```

```python
df_timeofday_artist_songsplayed = pd.DataFrame(timeofday_artist_songsplayed, columns=['Artist', 'Total_songs_played'])
df_timeofday_artist_songsplayed.insert(loc=0, column='Time_of_day', value=td)
```

# 8)	Favorite day of the week for listening songs based on

## b.	Total listening time


```python
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
favday_listentime = list()
favday_songsplayed = list()
for i in weekdays:
    temp = data[data['day_of_week'] == i]
    favday_listentime.append(temp['ms_played'].sum())
    favday_songsplayed.append(temp.shape[0])
```

# 9)	Songs that were most repeated in a single day


```python
dates_unique = data['date'].unique()
most_played_day = list()
for i in range(0,len(dates_unique)):
    most_played_day.append(['nothing', 0])
```

```python
for i in range(0,len(dates_unique)):
    temp = data[data['date'] == dates_unique[i]]
    temp_songs = temp['master_metadata_track_name'].unique()
    for j in temp_songs:
        temp2 = temp[temp['master_metadata_track_name'] == j]
        if temp2.shape[0] > most_played_day[i][1]:
            most_played_day[i][0] = j
            most_played_day[i][1] = temp2.shape[0]
```

```python
df = pd.DataFrame(most_played_day, columns=['Name_of_song', 'No_of_times_played'])
df.insert(loc=0, column='Date', value=dates_unique)
```

```python
df_most_played_singleday = df.sort_values(by=['No_of_times_played'], ascending=False).iloc[:50]
df_most_played_singleday.index = range(0, 50)
```
